```python
# Install required libraries (only needed in Colab)
!pip install transformers torch --quiet

# Import GPT-2 from Huggingface
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Set your prompt
prompt = "In the future, intelligent transportation systems will"

# Tokenize prompt
inputs = tokenizer.encode(prompt, return_tensors="pt")

# Generate text from model
outputs = model.generate(inputs, max_length=100, do_sample=True, temperature=0.8)

# Decode and print generated text
print("\nüìÑ Generated Text:\n")
print(tokenizer.decode(outputs[0], skip_special_tokens=True))

```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased")

inputs = tokenizer("I love using Hugging Face!", return_tensors="pt")
outputs = model(**inputs)

print("Logits:", outputs.logits)
```
